```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/ab31s/Downloads/ft_prod_specific_piecewise.csv")
```

```python

def plot_graph(df, x_col, y_col, graph_type='line', title='Graph', xlabel=None, ylabel=None):
   
    plt.figure(figsize=(12, 6))
    
    if graph_type == 'line':
        sns.lineplot(x=x_col, y=y_col, data=df)
    elif graph_type == 'bar':
        sns.barplot(x=x_col, y=y_col, data=df)
    elif graph_type == 'scatter':
        sns.scatterplot(x=x_col, y=y_col, data=df)
    elif graph_type == 'hist':
        sns.histplot(df[y_col], bins=20, kde=True)
    elif graph_type == 'box':
        sns.boxplot(x=x_col, y=y_col, data=df)
    elif graph_type == 'heatmap':
        sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    else:
        print("Invalid graph type. Use 'line', 'bar', 'scatter', 'hist', 'box', or 'heatmap'.")
        return

    plt.title(title)
    plt.xlabel(xlabel if xlabel else x_col)
    plt.ylabel(ylabel if ylabel else y_col)
    plt.grid(True)
    plt.show()
```

```python
plot_graph(df, 'Species', 'PetalLengthCm', graph_type='bar', title='Bar Chart', xlabel="Species", ylabel="Length")
```

```python
plot_graph(df, 'SepalWidthCm', 'PetalLengthCm', graph_type='scatter', title='Bar Chart', xlabel="Species", ylabel="Length")
```
