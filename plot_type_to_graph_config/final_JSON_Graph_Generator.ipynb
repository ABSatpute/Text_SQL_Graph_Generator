```python
import json
import pandas as pd
import numpy as np
from pandas.api.types import infer_dtype

def generate_chart_json(json_result, chart_type="bar", value_column=None, label_column=None, colors=None):
    """
    Converts a JSON dataset to a Chart.js compatible JSON format with support for multiple datasets.

    Args:
        json_result (str or list): Input data in JSON format (list of dicts or JSON string).
        chart_type (str): Type of chart to generate. Options: 'bar', 'line', 'pie'.
        value_column (str or list): Column name(s) to use for chart values. If not provided, all numeric columns are used.
        label_column (str): Column name to use for labels. If not provided, the first datetime or string column is used.
        colors (list): Optional list of colors to use for datasets.

    Returns:
        dict: JSON structure compatible with Chart.js configuration.

    Raises:
        ValueError: If no suitable label or value column is found.
    """
    
    # Load data into DataFrame
    if isinstance(json_result, str):
        data = json.loads(json_result)
    else:
        data = json_result
    df = pd.DataFrame(data)

    # Detect and convert column data types
    for col in df.columns:
        series = df[col]
        inferred = infer_dtype(series, skipna=True)

        if inferred in ["string", "unicode"]:
            df[col] = series.astype(str)
        elif inferred == "boolean":
            df[col] = series.astype(bool)
        elif inferred == "integer":
            if "timestamp" in col.lower() or "time" in col.lower():
                try:
                    df[col] = pd.to_datetime(series, unit="ms")
                    continue
                except Exception:
                    pass
            df[col] = pd.to_numeric(series, errors="coerce").astype("Int64")
        elif inferred == "floating":
            df[col] = pd.to_numeric(series, errors="coerce")
        else:
            try:
                df[col] = pd.to_datetime(series)
            except Exception:
                df[col] = series.astype(str)

    # Detect label column if not provided
    if not label_column:
        for col in df.columns:
            if pd.api.types.is_datetime64_any_dtype(df[col]) or pd.api.types.is_string_dtype(df[col]):
                label_column = col
                break
        if not label_column:
            raise ValueError("No suitable label column (datetime/string) found.")

    # Determine value columns
    if value_column is None:
        value_columns = df.select_dtypes(include=[np.number]).columns.tolist()
        if len(value_columns) == 0:
            raise ValueError("No suitable value column (numeric) found.")
    elif isinstance(value_column, str):
        value_columns = [value_column]
    else:
        value_columns = value_column

    # Extract labels
    labels = (
        df[label_column].dt.strftime("%Y-%m-%d")
        if pd.api.types.is_datetime64_any_dtype(df[label_column])
        else df[label_column].astype(str)
    ).tolist()

    # Define color palette
    default_colors = [
        "#FF6384", "#36A2EB", "#FFCE56", "#4BC0C0",
        "#9966FF", "#FF9F40", "#66FF66", "#FF66B2",
        "#3399FF", "#CCFF33", "#FF9933", "#33CCFF"
    ]
    palette = colors or default_colors

    # Build chart structure
    chart_json = {
        "chart": {
            "type": chart_type,
            "data": {
                "labels": labels,
                "datasets": []
            },
            "options": {
                "scales": {
                    "yAxes": [{"ticks": {"beginAtZero": True}}]
                }
            }
        }
    }

    # Pie chart can only have one dataset
    if chart_type == "pie":
        if len(value_columns) > 1:
            raise ValueError("Pie chart supports only one value column.")
        col = value_columns[0]
        values = df[col].fillna(0).tolist()
        chart_json["chart"]["data"]["datasets"].append({
            "label": col,
            "data": values,
            "backgroundColor": palette[:len(values)]
        })
    else:
        for idx, col in enumerate(value_columns):
            values = df[col].fillna(0).tolist()
            chart_json["chart"]["data"]["datasets"].append({
                "label": col.replace("_", " ").title(),
                "data": values,
                "backgroundColor": palette[idx % len(palette)],
                "borderColor": palette[idx % len(palette)],
                "borderWidth": 1
            })

    return chart_json
```

```python

import pandas as pd
import random

def get_chartjs_config_auto(data, chart_type="line", stacked=False):
    df = pd.DataFrame(data)
    
    for col in df.columns:
        series = df[col]
        inferred = infer_dtype(series, skipna=True)

        if inferred in ["string", "unicode"]:
            df[col] = series.astype(str)
        elif inferred == "boolean":
            df[col] = series.astype(bool)
        elif inferred == "integer":
            max_len = series.dropna().astype(str).str.len().max()
            if max_len is not None and max_len >= 13:  # Likely a timestamp in milliseconds
                # Convert to datetime if it's a timestamp
                # Assume UNIX timestamp in milliseconds
                try:
                    df[col] = pd.to_datetime(series, unit="ms")
                    continue
                except Exception:
                    pass
            df[col] = pd.to_numeric(series, errors="coerce").astype("Int64")
        elif inferred == "floating":
            df[col] = pd.to_numeric(series, errors="coerce")
        else:
            # Try datetime conversion, fallback to string
            try:
                df[col] = pd.to_datetime(series)
            except Exception:
                df[col] = series.astype(str)
    print(f"inferred data types: {df.dtypes}")

    # Auto-detect x_col and y_cols
    non_numeric = df.select_dtypes(exclude='number')
    numeric = df.select_dtypes(include='number')

    if non_numeric.empty:
        raise ValueError("No non-numeric columns found to use as x-axis labels.")

    x_col = non_numeric.nunique().idxmax()
    y_cols = numeric.columns.tolist()

    labels = df[x_col].astype(str).tolist()

    def random_color():
        return f"rgba({random.randint(50,200)}, {random.randint(50,200)}, {random.randint(50,200)}, 1)"

    datasets = []

    if chart_type == "pie":
        if len(y_cols) != 1:
            raise ValueError("Pie chart supports only one numeric column.")
        values = df[y_cols[0]].fillna(0).tolist()
        background_colors = [random_color() for _ in values]
        datasets = [{
            "label": y_cols[0],
            "data": values,
            "backgroundColor": background_colors
        }]
    else:
        for col in y_cols:
            datasets.append({
                "label": col,
                "data": df[col].fillna(0).tolist(),
                "borderColor": random_color(),
                "backgroundColor": random_color(),
                "fill": chart_type == "line" or chart_type == "bar",
                "stack": stacked
            })

    config = {
        "type": chart_type,
        "data": {
            "labels": labels,
            "datasets": datasets
        },
        "options": {
            "responsive": True,
            "scales": {
                "x": {
                    "stacked": stacked,
                    "title": {
                        "display": True,
                        "text": x_col
                    }
                },
                "y": {
                    "stacked": stacked,
                    "title": {
                        "display": True,
                        "text": " / ".join(y_cols)
                    }
                }
            },
            "plugins": {
                "title": {
                    "display": True,
                    "text": f"{chart_type.capitalize()} Chart of {' & '.join(y_cols)} Over {x_col}"
                }
            }
        }
    }

    if chart_type == "pie":
        # Pie chart doesn't use scales
        del config["options"]["scales"]

    return config
```

```python
import psycopg2
import json
from decimal import Decimal
from datetime import datetime, date, time
from uuid import UUID


class PostgresDataEncoder(json.JSONEncoder):
    """
    Custom JSON encoder to handle PostgreSQL native types.
    Converts datetime, date, time, UUID, Decimal into JSON-serializable types.
    """
    def default(self, obj):
        if isinstance(obj, (datetime, date, time)):
            return obj.isoformat()
        elif isinstance(obj, Decimal):
            return float(obj)
        elif isinstance(obj, UUID):
            return str(obj)
        return super().default(obj)


def fetch_data_from_postgres(query, db_config):
    """
    Fetches data from PostgreSQL and returns it as a JSON string.
    Native Python types are preserved (no conversion to string).
    """
    conn = None
    result_data = []

    try:
        # Connect to PostgreSQL
        conn = psycopg2.connect(**db_config)
        cur = conn.cursor()

        # Execute query
        print(f"\n🔎 Executing query:\n{query}\n")
        cur.execute(query)
        rows = cur.fetchall()

        # Fetch column names
        col_names = [desc[0] for desc in cur.description]

        # Convert rows to list of dictionaries
        for row in rows:
            result_data.append(dict(zip(col_names, row)))

    except psycopg2.Error as e:
        print(f"❌ Error: {e}")
        return json.dumps({"error": str(e)}, indent=4)

    finally:
        if conn:
            conn.close()

    # Use the custom encoder here ✅
    return json.dumps(result_data, indent=4, cls=PostgresDataEncoder)


# ✅ PostgreSQL connection config
db_config = {
    "host": "localhost",
    "database": "postgres",
    "user": "postgres",
    "password": "root",
    "port": "5432"
}

# 📥 Get the query from user
query = input("\n🛠️ Enter your SQL query: ")

# 📤 Fetch and display the result in JSON
json_result = fetch_data_from_postgres(query, db_config)
print("\n📊 Query Result in JSON format:")
print(json_result)
```

```python
import pandas as pd
import random
from pandas.api.types import infer_dtype

def get_chartjs_config_auto(json_result, chart_type="line"):
    if isinstance(json_result, str):
        data = json.loads(json_result)
    else:
        data = json_result
    df = pd.DataFrame(data)


    # Convert data types based on inference
    for col in df.columns:
        series = df[col]
        inferred = infer_dtype(series, skipna=True)

        if inferred in ["string", "unicode"]:
            df[col] = series.astype(str)
        elif inferred == "boolean":
            df[col] = series.astype(bool)
        elif inferred == "integer":
            max_len = series.dropna().astype(str).str.len().max()
            if max_len and max_len >= 13:
                try:
                    df[col] = pd.to_datetime(series, unit="ms")
                    continue
                except Exception:
                    pass
            df[col] = pd.to_numeric(series, errors="coerce").astype("Int64")
        elif inferred == "floating":
            df[col] = pd.to_numeric(series, errors="coerce")
        else:
            try:
                df[col] = pd.to_datetime(series)
            except Exception:
                df[col] = series.astype(str)

    non_numeric = df.select_dtypes(exclude='number')
    numeric = df.select_dtypes(include='number')
    datetime_cols = df.select_dtypes(include='datetime64')

    def random_color():
        return f"rgba({random.randint(50,200)}, {random.randint(50,200)}, {random.randint(50,200)}, 0.7)"

    datasets = []
    labels = []

    # PIE CHART
    if chart_type == "pie":
        if len(numeric.columns) != 1:
            raise ValueError("Pie chart supports only one numeric column.")
        y_col = numeric.columns[0]
        x_col = non_numeric.columns[0] if not non_numeric.empty else "Category"
        labels = df[x_col].astype(str).tolist()
        values = df[y_col].fillna(0).tolist()
        background_colors = [random_color() for _ in values]
        datasets = [{
            "label": y_col,
            "data": values,
            "backgroundColor": background_colors
        }]
        
    # === SCATTER CHART ===
    elif chart_type == "scatter":
        if numeric.shape[1] < 2:
            raise ValueError("Scatter chart requires at least two numeric columns.")
        x_col, y_col = numeric.columns[:2]
        data_points = [{"x": x, "y": y} for x, y in zip(df[x_col], df[y_col])]
        return {
            "type": "scatter",
            "data": {
                "datasets": [{
                    "label": f"{y_col} vs {x_col}",
                    "data": data_points,
                    "backgroundColor": random_color()
                }]
            },
            "options": {
                "responsive": True,
                "scales": {
                    "x": {"type": "linear", "title": {"display": True, "text": x_col}},
                    "y": {"title": {"display": True, "text": y_col}}
                },
                "plugins": {
                    "title": {
                        "display": True,
                        "text": f"Scatter Plot of {y_col} vs {x_col}"
                    }
                }
            }
        }

    
    # LINE CHART
    elif chart_type == "line" and not datetime_cols.empty:
        x_col = datetime_cols.columns[0]
        df = df.sort_values(x_col)
        y_cols = numeric.columns.tolist()
        group_col = None

        for col in non_numeric.columns:
            if col != x_col:
                group_col = col
                break

        if group_col:
            for y_col in y_cols:
                grouped = df[[x_col, group_col, y_col]].dropna()
                for group_name, group_df in grouped.groupby(group_col):
                    group_df = group_df.sort_values(x_col)
                    datasets.append({
                        "label": f"{y_col} - {group_name}",
                        "data": group_df[y_col].tolist(),
                        "borderColor": random_color(),
                        "backgroundColor": random_color(),
                        "fill": False
                    })
            labels = [str(val) for val in sorted(df[x_col].dropna().unique())]
        else:
            labels = df[x_col].astype(str).tolist()
            for y_col in y_cols:
                datasets.append({
                    "label": y_col,
                    "data": df[y_col].fillna(0).tolist(),
                    "borderColor": random_color(),
                    "backgroundColor": random_color(),
                    "fill": True
                })

    # BAR CHART (auto decide between simple and stacked)
    elif chart_type == "bar":
        y_col = numeric.columns[0] if not numeric.empty else None
        x_col = None
        group_col = None

        if not datetime_cols.empty:
            x_col = datetime_cols.columns[0]
        elif not non_numeric.empty:
            x_col = non_numeric.columns[0]

        # Check for second non-numeric column to enable stacked bar
        for col in non_numeric.columns:
            if col != x_col:
                group_col = col
                break

        if group_col and y_col:
            # STACKED BAR
            grouped = df[[x_col, group_col, y_col]].dropna()
            pivot = grouped.pivot_table(index=x_col, columns=group_col, values=y_col, aggfunc='sum').fillna(0)
            labels = [str(l) for l in pivot.index]

            for group_name in pivot.columns:
                datasets.append({
                    "label": str(group_name),
                    "data": pivot[group_name].tolist(),
                    "backgroundColor": random_color(),
                    "stack": "stack1"
                })

            stacked = True

        else:
            # SIMPLE BAR
            y_cols = numeric.columns.tolist()
            if not x_col:
                raise ValueError("No x-axis (non-numeric or datetime) column available for bar chart.")
            labels = df[x_col].astype(str).tolist()
            for col in y_cols:
                datasets.append({
                    "label": col,
                    "data": df[col].fillna(0).tolist(),
                    "backgroundColor": random_color(),
                })
            stacked = False

    else:
        raise ValueError(f"Unsupported chart type: {chart_type}")
    
    

    # Final Chart.js Config
    config = {
        "type": "bar" if chart_type == "bar" else chart_type,
        "data": {
            "labels": labels,
            "datasets": datasets
        },
        "options": {
            "responsive": True,
            "plugins": {
                "title": {
                    "display": True,
                    "text": f"{chart_type.capitalize()} Chart"
                }
            }
        }
    }

    # Axis Config for non-pie charts
    if chart_type != "pie":
        config["options"]["scales"] = {
            "x": {
                "stacked": stacked if chart_type == "bar" else False,
                "title": {"display": True, "text": x_col if 'x_col' in locals() else ""}
            },
            "y": {
                "stacked": stacked if chart_type == "bar" else False,
                "title": {"display": True, "text": " / ".join(numeric.columns.tolist())}
            }
        }

    return config
config = get_chartjs_config_auto(json_result, chart_type="line")
config
```

```python
import pandas as pd
import random
from pandas.api.types import infer_dtype

def get_config_auto(json_result, chart_type="line"):
    if isinstance(json_result, str):
        data = json.loads(json_result)
    else:
        data = json_result
    df = pd.DataFrame(data)

    # Type inference & conversion
    for col in df.columns:
        series = df[col]
        inferred = infer_dtype(series, skipna=True)

        if inferred in ["string", "unicode"]:
            df[col] = series.astype(str)
        elif inferred == "boolean":
            df[col] = series.astype(bool)
        elif inferred == "integer":
            max_len = series.dropna().astype(str).str.len().max()
            if max_len and max_len >= 13:
                try:
                    df[col] = pd.to_datetime(series, unit="ms")
                    continue
                except Exception:
                    pass
            df[col] = pd.to_numeric(series, errors="coerce").astype("Int64")
        elif inferred == "floating":
            df[col] = pd.to_numeric(series, errors="coerce")
        else:
            try:
                df[col] = pd.to_datetime(series)
            except Exception:
                df[col] = series.astype(str)

    print(f"inferred data types: {df.dtypes}")

    # Column type groups
    datetime_cols = df.select_dtypes(include="datetime")
    numeric = df.select_dtypes(include="number")
    non_numeric = df.select_dtypes(exclude="number")

    def random_color():
        return f"rgba({random.randint(50,200)}, {random.randint(50,200)}, {random.randint(50,200)}, 1)"

    # === PIE CHART ===
    if chart_type == "pie":
        if numeric.shape[1] != 1:
            raise ValueError("Pie chart requires exactly one numeric column.")
        x_col = non_numeric.nunique().idxmax()
        y_col = numeric.columns[0]
        labels = df[x_col].astype(str).tolist()
        values = df[y_col].fillna(0).tolist()
        return {
            "type": "pie",
            "data": {
                "labels": labels,
                "datasets": [{
                    "label": y_col,
                    "data": values,
                    "backgroundColor": [random_color() for _ in values]
                }]
            },
            "options": {
                "responsive": True,
                "plugins": {
                    "title": {
                        "display": True,
                        "text": f"Pie Chart of {y_col} Over {x_col}"
                    }
                }
            }
        }

    # === SCATTER CHART ===
    if chart_type == "scatter":
        if numeric.shape[1] < 2:
            raise ValueError("Scatter chart requires at least two numeric columns.")
        x_col, y_col = numeric.columns[:2]
        data_points = [{"x": x, "y": y} for x, y in zip(df[x_col], df[y_col])]
        return {
            "type": "scatter",
            "data": {
                "datasets": [{
                    "label": f"{y_col} vs {x_col}",
                    "data": data_points,
                    "backgroundColor": random_color()
                }]
            },
            "options": {
                "responsive": True,
                "scales": {
                    "x": {"type": "linear", "title": {"display": True, "text": x_col}},
                    "y": {"title": {"display": True, "text": y_col}}
                },
                "plugins": {
                    "title": {
                        "display": True,
                        "text": f"Scatter Plot of {y_col} vs {x_col}"
                    }
                }
            }
        }

    # === BAR or LINE ===
    # Determine x_col: prefer datetime, fallback to most diverse non-numeric
    if not datetime_cols.empty:
        x_col = datetime_cols.columns[0]
    elif not non_numeric.empty:
        x_col = non_numeric.nunique().idxmax()
    else:
        raise ValueError("No suitable x-axis column found.")

    y_cols = numeric.columns.tolist()
    labels = df[x_col].astype(str).tolist()

    # Detect stack condition: only for bar, if a non-numeric category column exists
    stack = chart_type == "bar" and not non_numeric.empty

    datasets = []
    for col in y_cols:
        datasets.append({
            "label": col,
            "data": df[col].fillna(0).tolist(),
            "borderColor": random_color(),
            "backgroundColor": random_color(),
            "fill": chart_type == "line" or chart_type == "bar",
            "stack": stack
        })

    return {
        "type": chart_type,
        "data": {
            "labels": labels,
            "datasets": datasets
        },
        "options": {
            "responsive": True,
            "scales": {
                "x": {
                    "stacked": stack,
                    "title": {"display": True, "text": x_col}
                },
                "y": {
                    "stacked": stack,
                    "title": {"display": True, "text": " / ".join(y_cols)}
                }
            },
            "plugins": {
                "title": {
                    "display": True,
                    "text": f"{chart_type.capitalize()} Chart of {' & '.join(y_cols)} Over {x_col}"
                }
            }
        }
    }
```
